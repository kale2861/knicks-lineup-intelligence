```python
import pandas as pd
import numpy as np
from pathlib import Path

PROCESSED_DIR = Path("../data/processed")

df = pd.read_csv(PROCESSED_DIR / "knicks_lineups_clean.csv")
```

## Lineup Features

```python
df["lineup_quality"] = pd.cut(
    df["net_rating"],
    bins=[-999, -5, 0, 5, 999],
    labels=["Negative", "Neutral", "Positive", "Elite"]
)

df["offense_tier"] = pd.cut(
    df["off_rating"],
    bins=[0, 105, 115, 125, 999],
    labels=["Poor", "Average", "Strong", "Elite"]
)

df["defense_tier"] = pd.cut(
    df["def_rating"],
    bins=[0, 105, 112, 120, 999],
    labels=["Elite", "Strong", "Average", "Poor"]
)
```

## Stability Score

```python
df["lineup_stability_score"] = (
    df["min"] / df["min"].sum()
).round(4)
```

## Balance score

```python
df["two_way_score"] = (
    df["off_rating"] - df["def_rating"]
).round(2)
```

```python
df.to_csv(
    PROCESSED_DIR / "knicks_lineups_features.csv",
    index=False
)
```
