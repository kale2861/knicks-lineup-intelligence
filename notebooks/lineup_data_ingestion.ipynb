```python
from nba_api.stats.endpoints import leaguedashlineups
import pandas as pd
from pathlib import Path

RAW_DIR = Path("../data/raw")
PROCESSED_DIR = Path("../data/processed")

RAW_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

## Pull Knicks lineup data

```python
KNICKS_TEAM_ID = "1610612752"

lineups = leaguedashlineups.LeagueDashLineups(
    season="2025-26",
    team_id_nullable=KNICKS_TEAM_ID,
    measure_type_detailed_defense="Advanced",
    per_mode_detailed="Per100Possessions"
)

df = lineups.get_data_frames()[0]

df.head()
```

## Save Raw Data

```python
df.to_csv(
    RAW_DIR / "knicks_lineups_raw_2025_26.csv",
    index=False
)
```

## Clean and Filter Lineups

```python
lineups_clean = df.copy()

lineups_clean.columns = (
    lineups_clean.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
```

## Filter Out Sample Lineups

```python
lineups_clean = lineups_clean[
    lineups_clean["min"] >= 25
].copy()
```

```python
lineups_clean.to_csv(
    PROCESSED_DIR / "knicks_lineups_clean.csv",
    index=False
)
```
